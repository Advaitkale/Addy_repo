**PRE-PROCESSING OF DATA**
1. Convert the 'Status' Column values into integer (0 or 1)
2. Remove Null and NaN values in the dataset


3. Features =[['GRE Total','TOEFL','Paper','CGPA','Work Ex']],

   Target = ['Status']


4. Converting CGPA Score to only out of 4.
5. Remove 'Mo' in "Work Ex" Column values.
6. Normalizing all the features
7. Growth Trend of Students getting admitted in a particular term


```
for i in range(len(admit_data['Status'])):
  if(admit_data['Status'][i] in ['Admit','admit','ADMIT']):
    admit_data['Status'].iloc[i]=1
  elif(admit_data['Status'][i] in ['Reject','reject','REJECT']):
    admit_data['Status'].iloc[i]=0

print(admit_data.head())

print("*******************************************")
print(admit_data['Status'])
```

```
for i in range(len(admit_data['CGPA'])):
  if(admit_data['CGPA'].iloc[i]>4 and admit_data['CGPA'].iloc[i]<=10):
    admit_data['CGPA'].iloc[i]=admit_data['CGPA'].iloc[i]*0.4
  elif(admit_data['CGPA'].iloc[i]>10 and admit_data['CGPA'].iloc[i]<=100):
    admit_data['CGPA'].iloc[i]=admit_data['CGPA'].iloc[i]*0.04
  else:
    admit_data['CGPA'].iloc[i]=admit_data['CGPA'].iloc[i]*1

admit_data
```

```
#admit_data['Work Ex']=admit_data['Work Ex'].map(lambda x: x.rstrip(' Mo'))
for i in range(len(admit_data['Work Ex'])):
  l=str(admit_data['Work Ex'][i]).split(' ')
  admit_data['Work Ex'][i]=l[0]
admit_data.head()
```

```
for i in range(len(admit_data['Work Ex'])):
  if(str(admit_data['Work Ex'].iloc[i]) in ['N/A','nan','NaN']):
    admit_data['Work Ex'].iloc[i]=0
admit_data
```

```
admit_data['Term']=admit_data['Term'].fillna('Unknown')
count=0
d={}
a=list(pd.unique(admit_data['Term']))
for i in range(len(a)):
  for j in range(len(admit_data['Term'])):
    if(a[i]==admit_data['Term'][j]):
      count+=1
  #print(a[i],"\t",count,'\n')
  d[a[i]]=count

  count=0

s=pd.DataFrame()

s['Term'] = d.keys()
s['No. of Students'] =d.values()


plt.plot(s['Term'],s['No. of Students'])
plt.title('Trend of Students getting admitted every term')
plt.ylabel('No. of Students')
plt.xticks(s['Term'],rotation=90)
plt.xlabel('Term')
plt.show()
```

```
admit_data['Students Admitted in Particular Term']=admit_data['CGPA']
for i in range(len(s)):
  for j in range(len(admit_data)):
    if(s['Term'][i]==admit_data['Term'][j]):
      admit_data['Students Admitted in Particular Term'][j]=s['No. of Students'][i]


admit_data.head()
```

```

a=1

b=1

c=1

e=1


for i in range(len(admit_data)):
  admit_data['GRE Total'].iloc[i]=(admit_data['GRE Total'].iloc[i])/(a)
  admit_data['TOEFL'].iloc[i]=(admit_data['TOEFL'].iloc[i])/(b)
  admit_data['CGPA'].iloc[i]=(admit_data['CGPA'].iloc[i])/(c)
  
  admit_data['Students Admitted in Particular Term'].iloc[i]=(admit_data['Students Admitted in Particular Term'].iloc[i])/(e)

admit_data.head()
```

```
admit_data['University']=admit_data['University'].fillna('Unknown')
admit_data['Target Major']=admit_data['Target Major'].fillna('Unknown')
score={}
m=[]
l=[]
c=1
for i in range(len(uni)):
  for j in range(len(admit_data)):
    if(uni[i]==admit_data['University'][j]):
      l.append(admit_data['Target Major'][j])
      l.append(c)
      m.append(l)
      c+=1
    l=[]
    
    
  score[uni[i]]=m
  m=[]
  
score
```

```
admit_data['University Code']=admit_data['CGPA']
for i in range(len(uni)):
  for j in range(len(admit_data)):
    if(uni[i]==admit_data['University'][j]):
      for k in range(len(score[uni[i]])):
        if(score[uni[i]][k][0]==admit_data['Target Major'][j]):
          admit_data['University Code'][j]=score[uni[i]][k][1]
admit_data
```

***FEATURE ENGINEERING STAGE***

```
X=admit_data[['GRE Total','TOEFL','Papers','CGPA','Work Ex','University Rank']]
y = admit_data['Status']
```

```
X
```

```
type(admit_data['Work Ex'][0])
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y,random_state=42,test_size=0.1)
```

```
X_test
```

```
X_train
```

```
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

#Create a Gaussian Classifier
gnb = GaussianNB().fit(X_train,y_train)

xgb = XGBClassifier().fit(X_train, y_train)

classifier = LogisticRegression(random_state = 0).fit(X_train, y_train) 

clf=RandomForestClassifier(n_estimators=3000).fit(X_train,y_train)

knn = KNeighborsClassifier(n_neighbors=100).fit(X_train, y_train) 

dtc=DecisionTreeClassifier(max_depth=1000).fit(X_train, y_train)
 
y_pred = classifier.predict_proba(X_test)[:, 1]
y_prediction=classifier.predict(X_test)

y1_pred=clf.predict_proba(X_test)[:,1]
y1_prediction=clf.predict(X_test)

y2_pred=gnb.predict_proba(X_test)[:,1]
y2_prediction=gnb.predict(X_test)

y3_pred=xgb.predict_proba(X_test)[:,1]
y3_prediction=xgb.predict(X_test)

y4_pred=knn.predict_proba(X_test)[:,1]
y4_prediction=knn.predict(X_test)

y5_pred=dtc.predict_proba(X_test)[:,1]
y5_prediction=dtc.predict(X_test)


print('Logistic Regression')
print('*******************************************************************')
print(accuracy_score(y_test,y_prediction))
print(classifier.coef_)
print('*******************************************************************')

print('Random Forest Classifier')

print(accuracy_score(y_test,y1_prediction))
print('*******************************************************************')

print('Gaussian Naive Bayes Classifier')

print(accuracy_score(y_test,y2_prediction))
print('*******************************************************************')

print('XGBoost Classifier')

print(accuracy_score(y_test,y3_prediction))
print('*******************************************************************')

print('k-NN Classifier')

print(accuracy_score(y_test,y4_prediction))
print('*******************************************************************')

print('Decision Tree Classifier')

print(accuracy_score(y_test,y5_prediction))
print('*******************************************************************')
```
